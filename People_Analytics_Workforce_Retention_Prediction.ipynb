# TASK #3: VISUALIZE DATASET

```python
# Let's replace 'Attritition' , 'OverTime' , 'Over18' column with integers before performing any visualizations
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x:1 if x == 'Yes' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x:1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x:1 if x == 'Y' else 0)
```

```python
# checking the dataset again after changes
employee_df[['Attrition','OverTime','Over18']].head()
```

```python
# Let's see if we have any missing data
employee_df.isnull().sum()
```

```python
# we have checked about missing data and we don't have any missing data but we will check it with visualization too
plt.figure(figsize=(10,4))
sns.heatmap(employee_df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# we  are using a histogram to check all data of a dataset for more info visualy together
employee_df.hist(bins=30,figsize=(20,20))

# Several features such as 'MonthlyIncome' and 'TotalWorkingYears' are tail heavy
```

```python
# It makes sense to drop 'EmployeeCount' , 'Standardhours' and 'Over18' since they do not change from one employee to the other
# Let's drop 'EmployeeNumber' as well
employee_df.drop(['EmployeeCount','StandardHours','Over18','EmployeeNumber'], axis=1,inplace=True)
```

# TASK #4: CREATE TESTING AND TRAINING DATASET & PERFORM DATA CLEANING

```python
# Checking data again for cleaning
employee_df.head(3)
```

```python
# Checking all categorical data
X_cat = employee_df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
```

```python
X_cat
# we need to change this categorical data to numbers
```

```python
# importing important library to change categorical data to number
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
# checking the changes in data
X_cat.shape
```

```python
# converting this new data to a dataframe
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat
```

```python
# note that we dropped the target 'Atrittion'
X_numerical = employee_df[['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
# doing the Concatenation the all categorical and numerical dataframe together
X_all = pd.concat([X_cat,X_numerical],axis=1)
```

```python
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# X_all is a DataFrame, convert its column names to strings before applying MinMaxScaker
X_all.columns = X_all.columns.astype(str)

X = scaler.fit_transform(X_all)
```

```python
# X
```

```python
# now we are putting our traget column in y variable from our employee dataset
y = employee_df['Attrition']
y
```

# TASK #5: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split

X_tarin, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=101)
```

```python
# checking the tarin test split shape
X_tarin.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_tarin,y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
# Testing Set Performance
y_predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# we have used .2f for only 2 decimal places
print(f'Accuracy {100 * accuracy_score(y_predictions, y_test):.2f} %')
```

```python
cm = confusion_matrix(y_predictions,y_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predictions))
```

# TASK #6: TRAIN AND EVALUATE A RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_tarin,y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
# Testing Set Performance
cm = confusion_matrix(y_test,y_predictions)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_predictions,y_test))
```

# TASK #7: TRAIN AND EVALUATE A DEEP LEARNING MODEL

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(50,)))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics= ['accuracy'])
```

```python
epochs_hist = model.fit(X_tarin, y_train, epochs=100, batch_size=50)
```

```python
# oversampler = SMOTE(random_state=0)
# smote_train, smote_target = oversampler.fit_sample(X_train, y_train)
# epochs_hist = model.fit(smote_train, smote_target, epochs = 100, batch_size = 50)

y_predictions = model.predict(X_test)
y_predictions = (y_predictions >0.5)
```

```python
y_predictions
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
plt.plot(epochs_hist.history['accuracy'])
plt.title('Model Accuracy Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Accuracy')
plt.legend(['Training Accuracy'])
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predictions)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predictions))
```
